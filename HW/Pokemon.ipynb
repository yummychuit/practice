```python
class Pokemon:
    def __init__(self, level):
        self.level = level
        self.hp = self.level * 10
        
    def set_hp(self, point):
        self.hp += point
        
    def check_status(self): #죽었는지 살았는지
        if self.hp > 0:
            return True
        else: 
            return False
        
    def print_status(self):
        print(f'{self.name} 의 체력: {round(self.hp ,2)}')
```

```python
import random
class Pikachu(Pokemon):  
    name = '피카츄'
    types = ('elec')
    def body_attack(self, enemy): #피카츄가 공격한다! -> 객체 self 넣어줌
        enemy.set_hp(-1 * self.level)
        
    def thousand_volt(self, enemy):
        if 'water' in enemy.types:
            enemy.set_hp(-1.3 * self.level)
            print('효과는 굉장했다!')
        elif 'grass' in enemy.types:
            enemy.set_hp(-0.75 * self.level)
            print('효과가 별로인 것 같다...')
        elif 'elec' in enemy.types:
            enemy.set_hp(-0.6 * self.level)
            print('효과가 별로인 것 같다...')
        else:
            enemy.set_hp(-1.05 * self.level)
            
    def attack(self, enemy): 
        c = random.choice([1, 2])
        if c == 1:
            self.body_attack(enemy)
        else:
            self.thousand_volt(enemy)


class Squirtle(Pokemon):
    name = '꼬부기'
    types = ('water')
    def body_attack(self, enemy):
        enemy.set_hp(-0.98 * self.level)
    
    def water_attack(self, enemy):
        if 'fire' in enemy.types:
            enemy.set_hp(-1.5 * self.level)
            print('효과는 굉장했다!')
        elif 'grass' in enemy.types:
            enemy.set_hp(-0.7 * self.level)
            print('효과가 별로인 것 같다...')
        elif 'water' in enemy.types:
            enemy.set_hp(-0.6 * self.level)
            print('효과가 별로인 것 같다...')
        else:
            enemy.set_hp(-1.08 * self.level)
    
    def attack(self, enemy):
        c = random.choice([1, 2])
        if c == 1:
            self.body_attack(enemy)
        else:
            self.water_attack(enemy)
        

class Bulbasaur(Pokemon):
    name = '이상해씨'
    types = ('grass')
    def body_attack(self, enemy):
        enemy.set_hp(-1.05 * self.level)
        
    def growl_attack(self, enemy):
        if 'water' in enemy.types:
            enemy.set_hp(-1.55 * self.level)
            print('효과는 굉장했다!')
        elif 'fire' in enemy.types:
            enemy.set_hp(-0.77 * self.level)
            print('효과가 별로인 것 같다...')
        elif 'grass' in enemy.types:
            enemy.set_hp(-0.68 * self.level)
            print('효과가 별로인 것 같다...')
        else:
            enemy.set_hp(-0.95 * self.level)
            
    def attack(self, enemy):
        c = random.choice([1, 2])
        if c == 1:
            self.body_attack(enemy)
        else:
            self.growl_attack(enemy)
            

class Charmander(Pokemon):
    name = '파이리'
    types = ('fire')
    def scratch_attack(self, enemy):
        if 'grass' in enemy.types:
            enemy.set_hp(-1.45 * self.level)
            print('효과는 굉장했다!')
        elif 'fire' in enemy.types:
            enemy.set_hp(-0.65 * self.level)
            print('효과가 별로인 것 같다...')
        elif 'water' in enemy.types:
            enemy.set_hp(-0.55 * self.level)
            print('효과가 별로인 것 같다...')
        else:
            enemy.set_hp(-1.1 * self.level)
    
    def growl_attack(self, enemy):
        enemy.set_hp(-0.9 * self.level)
    
    def attack(self, enemy):
        c = random.choice([1, 2])
        if c == 1:
            self.scratch_attack(enemy)
        else:
            self.growl_attack(enemy)

```

```python
#
import random
class match_making:

    def my_pokemon():
        poke_list = ['Pikachu', 'Squirtle', 'Bulbasaur', 'Charmander']
        my_pok = random.choice(poke_list)
        return my_pok
        
    def enemy_pokemon():
        poke_list = ['Pikachu', 'Squirtle', 'Bulbasaur', 'Charmander']
        enemy_pok = random.choice(poke_list)
        return enemy_pok

functions = {
        'Pikachu': Pikachu,
        'Squirtle': Squirtle,
        'Bulbasaur': Bulbasaur,
        'Charmander': Charmander
    }        
functions[match_making.enemy_pokemon()]
functions[match_making.my_pokemon()]
```

```python
def battle(p1, p2):
    import random
    # 선공 후공 결정
    vs = [p1, p2]
    random.shuffle(vs)
    
    winner = 0
    prev, after = tuple(vs)
    while prev.check_status() and after.check_status():
        if prev.check_status:
            prev.attack(after)
            prev.print_status()
        else:
            winner = after
            break
        if after.check_status():
            after.attack(prev)
            after.print_status()
        else:
            winner = prev
            break
            
    print(f'{winner.name} 의 승리!')
    
a = functions[match_making.enemy_pokemon()]     
a_level = random.randrange(9, 30)
b = functions[match_making.my_pokemon()]
b_level = random.randrange(9, 30)


print(f'야생의 {b.name} 이/가 나타났다!')
print(f'{a.name} 너로 정했다!!\n')

battle(a(a_level), b(b_level))
```
